```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.array([12,24,36,48,60,72,84])
y = np.array([70,68,81,35,10,49,50])
z = np.array([9,18,27,36,48,54,63])
```

### Line Plot

```python
plt.title('line plot')
plt.xlabel(' x values ')
plt.ylabel(' y values ' )
plt.plot(x, y, marker = 'o', mfc = 'black', ls ='-.')
plt.plot(x, z, marker = 'o', mfc = 'black', ls ='-.')
```

```python
plt.title('line plot')
plt.xlabel('fisrt value (x)')
plt.ylabel('second value (z)')

plt.plot(x, z, marker = 'o', color = 'blue', mfc = 'r', ls = '--', ms = 7, linewidth = 1) # assigning marker, line color , line style

# ms = marker size
# ls =line stlye
# mfc = marker font color

plt.grid()
```

### Subplot

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.title('line plot')
plt.xlabel('fisrt value (x)')
plt.ylabel('second value (z)')

plt.plot(x, z, marker = 'o', color = 'blue', mfc = 'r', ls = '--', ms = 7, linewidth = 1) # assigning marker, line color , line style

# ms = marker size
# ls =line stlye
# mfc = marker font color
plt.grid()

#-------------------------------------------------------
plt.subplot(2,2,2)
plt.title('line plot')
plt.xlabel(' x values ')
plt.ylabel(' y values ' )
plt.plot(x, y, marker = 'o', mfc = 'black', ls ='-.')
plt.plot(x, z, marker = 'o', mfc = 'black', ls ='-.')
plt.grid()
```

### Bar Graph

```python
students = [56, 68, 46, 51, 65]

courses= ['java', 'python' , 'R', 'kotlin', 'dart']
```

```python
plt.title('students enrolled in courses')
plt.xlabel('Courses')
plt.ylabel('No. of Students')
plt.bar(courses, students,
        color = 'green', 
        edgecolor = 'yellow', 
        width= 0.7, #bar width
        linewidth = 1.5, #edge width
        alpha = 0.8)# opacity of the color
```

### Scatter Plot

```python
plt.scatter(x, y, 
            marker = '.');
plt.grid()
```

### Pie Charts

```python
plt.title('Pie Chart')
plt.pie(students,
        labels = courses,
        autopct = '%.2f%%',# auto assings percentage
        radius = 0.8, # size of pie chart
        labeldistance = 1.1, #distance of label from piechart
        colors = ['pink', 'orange', 'orange', 'skyblue', 'beige'],
        startangle = 290, #change angle of pie chart
        explode = [0, 0.2, 0, 0, 0], # highlight a particular portion
        frame = True,
       );
```

```python
plt.figure(figsize = (6,6))

plt.subplot(2,2,1)
plt.pie(students,
        labels = courses,
        autopct = '%.2f%%',# auto assings percentage
        radius = 0.8, # size of pie chart
        labeldistance = 1.1, #distance of label from piechart
        colors = ['pink', 'orange', 'orange', 'skyblue', 'beige'],
        startangle = 290, #change angle of pie chart
        explode = [0, 0.2, 0, 0, 0], # highlight a particular portion
       );

plt.subplot(2,2,2)
plt.xlabel(' x values ')
plt.ylabel(' y values ' )
plt.plot(x, y, marker = 'o', mfc = 'black', ls ='-.')
plt.plot(x, z, marker = 'o', mfc = 'black', ls ='-.')

plt.subplot(2,2,3)
plt.xlabel('Courses')
plt.ylabel('No. of Students')
plt.bar(courses, students,
        color = 'green', 
        edgecolor = 'yellow', 
        width= 0.7, #bar width
        linewidth = 1.5, #edge width
        alpha = 0.8)# opacity of the color

plt.subplot(2,2,4)
plt.pie(students,
        labels = courses,
        autopct = '%.2f%%',# auto assings percentage
        radius = 0.8, # size of pie chart
        labeldistance = 1.1, #distance of label from piechart
        colors = ['pink', 'orange', 'orange', 'skyblue', 'beige'],
        startangle = 290, #change angle of pie chart
        explode = [0, 0.2, 0, 0, 0], # highlight a particular portion
       );

plt.tight_layout() #automatically adjusts the padding
```

## Box Plot

```python
marks = [61,63,67,69,62,64,70,50,85,71]
```

```python
plt.boxplot(marks);
plt.grid()
```

```python
q3 = np.percentile(marks, 75)
q1 = np.percentile(marks, 25)
q3,q1
```

```python
iqr = q3 - q1 # interquartile range
iqr
```

```python
#lower bound
lb = q1 - 1.5 * iqr

#upper bound
ub = q3 + 1.5 * iqr
```

```python
lb, ub
```

# Seaborn

```python
tips = sns.load_dataset('tips')
tips
```

```python
#to remove the styles we need to assign deafault parameters
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
sns.jointplot(x = 'tip', y = 'total_bill',
              data = tips, hue = 'time');
plt.show()
```

```python
#histogram=freq distribution
plt.xlabel('tip amt')
plt.ylabel('freq')
sns.histplot(tips['tip'],kde=True) #kde is freq distribution curve
plt.show()
```

```python
#histogram=freq distribution
plt.xlabel('Day')
plt.ylabel('tip')
#sns.histplot(tips['tip'],kde=True)
sns.boxenplot(x='day',y='tip',data=tips)
plt.show()

#The boxen plot or letter value plot is similar to a box plot but provides more information about the shape of the distribution, particularly in the tails
```

```python
sns.pairplot(tips,hue='smoker')
plt.show()
#pairplot is similar to jointplot but
#1) scatter graph is generated for all the numeric values plkresent in the dataset
#2)diagonal graphs are the histogram
```
