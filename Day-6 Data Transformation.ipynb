```python
x = np.array([11,45,23,14,52,7,33,100])
```

```python
def normalize (a):
    norm = (a - a.min()) / (a.max() - a.min())
    return norm
```

```python
nx = normalize (x)
nx
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.plot(x)
plt.title('Original Data')
plt.grid()

plt.subplot(1,2,2)
plt.plot(nx)
plt.grid()
plt.title('Normalized Data')
```

```python
y = np.array([324.352,546,325,124,664,745,235,782])
```

```python
ny = normalize(y)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.plot(y)
plt.grid()
plt.title('Original Data')

plt.subplot(1,2,2)
plt.plot(ny)
plt.grid()
plt.title('Normalized Data')
```
