```python
df3=pd.read_csv('datasets/Mall_Customers.csv')
```

```python
df3
```

```python
x=df3.iloc[:,[3,4]]
x
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3,random_state=0)
labels=km.fit_predict(x)
labels
```

```python
km.inertia_ #sum squared (low error)
```

```python
s=[]
for i in range(1,20):
    km=KMeans(n_clusters=i)
    km.fit_predict(x)
    s.append(km.inertia_)
```

```python
s
```

```python
mp.xticks(range(1,20))
mp.plot(range(1,20),s,marker='o',mfc='r')
mp.grid()
```
