### 13. Plot the monthwise total count of shares for photo posts.

```python
share = df[df['Type'] == 'Photo'].groupby('Post Month')
share['share'].sum()
```

```python
phot = df[df['Type'] == 'Photo']
post_monthwise = phot[['Post Month', 'share']]
post_monthwise
```

```python
post_monthwise.groupby('Post Month')['share'].sum().plot(kind='bar')
plt.xlabel('Shares')
plt.title('Monthwise total count of shares for photo posts')
```
